```python
x = list(range(-10, 10))
print(x)
```

```python
y = []
for i in x:
    y.append(2*i)
print(x)
```

```python
plt.plot(x, y, 'ro')
plt.axis([-20, 20, -20, 20])
plt.title("y=2x")
plt.show()
```

##  GDP

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [67.0, 80.0, 257.0, 1686.0, 6505.0, 11863.0, 22105.3]
```

```python
plt.plot(years, gdp, color = 'green', marker = 'o', linestyle = 'solid')
plt.title('GDP per capita')
plt.ylabel('dollars')
plt.savefig("gdp_per_capita.png", dpi=600)
plt.show()
```
